```python
#importing required libraries and modules
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
#creating data sets

X, y = make_classification(n_samples=150, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,class_sep=3)
```

```python
#visualising generated data
plt.figure(figsize=(6,3))
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))  #creating sigmoid function to get probablity(y=1/x)

def perceptron_sigmoid(X,y):
    X=np.insert(X,0,1,axis=1) # creating array of feauture with one in first column
    learn_rate=0.1            #learning rate
    weights=np.ones(X.shape[1]) #initialising learning rate for first step
    
    for i in range(1000):
        j=np.random.randint(0,X.shape[0])
        y_hat=sigmoid(np.dot(weights,X[j]))
        weights=weights+learn_rate*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]
```

```python
intercept_,coef_=perceptron_sigmoid(X,y)
```

```python
#re writing our estimates in form of y=mx+b to get straight line
x_axis=np.linspace(-3,3,X.shape[0])
y_axis=-(intercept_/coef_[1])-(coef_[0]/coef_[1])*x_axis
plt.plot(x_axis,y_axis)
plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

Well the line obtained by using sigmoid inside perceptron improved our line compared to our previous discussion of perceptron

```python
#importing required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
#using make classification generate data set for classification problem
X, y = make_classification(n_samples=150, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,class_sep=3)
X[:3,:]
```

```python
plt.figure(figsize=(6,3))
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
def step(z):
    return 1 if z>0 else 0  # TO PREDICT THE POINT AS POSITIVE CLASS OR NOT 

def perceptron (X,y):
    X=np.insert(X,0,1,axis=1)  #creating x matrix where first column is full of 1 and rest are feautures
    weights=np.ones(X.shape[1])  #initialise the weights for initial step
    lr=0.1                     #learning rate to control the shift in line
    
    for i in range(1000):
        j=np.random.randint(0,X.shape[0])
        y_hat=step(np.dot(X[j],weights))
        weights=weights+lr*(y[j]-y_hat)*X[j]
    return weights[0],weights[1:]
        
```

```python
intercept_,coef_=perceptron(X,y)
print(intercept_)
print(coef_)
```

lets redefine our straight line for ploting purpose (to see visually how our line obtained using perceptron trick classifies data points)

```python
slope=-(coef_[0]/coef_[1])
y_intercept=-(intercept_/coef_[1])
x=np.linspace(-3,3,150)
y_axis=slope*x+y_intercept
```

```python
plt.figure(figsize=(6,4))
plt.plot(x,y_axis)
plt.scatter(X[:,0],X[:,1],c=y)
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

Now if observation is class 0 then its z value obtained by using straight line equation will be negative and if class is 1 then its z value will be greater than zero that is positive.

consider a observation whose x1 is -2 and x2 is -3 surely it is in purple group that is class 0 now letsd see what z value we get if substitue x1 and x2.

```python
z=intercept_+np.dot(np.array([-2,-3]),coef_)
z
```
